```python
import numpy as np

def estimate_n_sphere_volume(N_dim, N_samples, rng):

    # 1. Generate N_samples points, each with N_dim dimensions.
    #    The points are uniformly distributed within the bounding
    #    hypercube from [-1, 1] for each dimension.
    #    The shape of the array will be (N_samples, N_dim).
    points = rng.uniform(-1, 1, size=(N_samples, N_dim))
    
    # 2. Calculate the squared distance from the origin for all points.
    #    Sum the squares along the dimension axis (axis=1).
    r_squared = np.sum(points**2, axis=1)
    
    # 3. A point is inside the unit sphere if its squared distance
    #    is less than or equal to 1 (since r^2 = 1^2 = 1).
    #    Count how many such points exist.
    num_inside = np.sum(r_squared <= 1)
    
    # 4. The volume of the bounding hypercube is (2^N_dim)
    #    (since each of the N_dim sides has length 2).
    volume_hypercube = 2**N_dim
    
    # 5. The ratio of (points inside / total points) is approximately
    #    the ratio of (sphere volume/hypercube volume).
    #
    #    Volume_Sphere = (num_inside / N_samples) * Volume_Hypercube
    volume_sphere_estimate = (num_inside / N_samples) * volume_hypercube
    
    return volume_sphere_estimate

def main():

    rng = np.random.default_rng()
    
    # Number of samples for each dimension.
    # Higher dimensions require more samples for accuracy
    # as the sphere's volume becomes a tiny fraction
    # of the hypercube's volume.
    N_SAMPLES = 1_000_000
    MAX_DIM = 10
    
    volumes = []
    
    print(f"Estimating N-sphere volumes with {N_SAMPLES:,} samples per dimension:\n")
    
    for N in range(1, MAX_DIM + 1):
        volume = estimate_n_sphere_volume(N, N_SAMPLES, rng)
        volumes.append(volume)
        print(f"Dimension {N:2}: Estimated Volume = {volume:.6f}")
        
    max_volume = max(volumes)
    # Add 1 because the list is 0-indexed (N=1 is at index 0)
    max_volume_dim = volumes.index(max_volume) + 1 
    
    print(f"The highest volume ({max_volume:.6f}) was found at N = {max_volume_dim} dimensions.")

if __name__ == "__main__":
    main()
```

```python
# 7.251

import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

def g(x):
    # Use np.where to handle x=0, which would be 0 * inf = nan
    # (x**1.5 + x**0.5) is 0 at x=0.
    term1 = x**1.5 + x**0.5
    term2 = np.exp(-x)
    return np.where(x == 0, 0.0, term1 * term2)

def f(x):
    return x**1.5 + x**0.5

def plot_integrand():
    x_plot = np.linspace(0, 10, 500)
    y_plot = g(x_plot)
    
    fig, ax = plt.subplots()
    ax.plot(x_plot, y_plot, label=r'$g(x) = \frac{x^2+x}{\sqrt{x}} e^{-x}$')
    ax.set_title("Integrand for (7.251)")
    ax.set_xlabel("x")
    ax.set_ylabel("g(x)")
    ax.grid(True)
    ax.legend()
    plt.show()

def run_monte_carlo(N_samples, rng):

    print(f"\nRunning Monte Carlo with {N_samples:,} samples")
    
    # 1. Draw N_samples from U(0, 1)
    u_samples = rng.uniform(0, 1, size=N_samples)
    
    # 2. Apply inversion x = -ln(u) to get Exponential(1) samples.
    #    p(x) = exp(-x)
    x_samples = -np.log(u_samples)
    
    # 3. Calculate f(x) for all samples
    f_values = f(x_samples)
    
    # 4. The estimate is the mean of f(x)
    estimate = np.mean(f_values)
    
    return estimate

def main():
    # 1. Plot
    plot_integrand()
    
    # 2. Integrate
    rng = np.random.default_rng()
    N_SAMPLES = 2_000_000
    
    mc_estimate = run_monte_carlo(N_SAMPLES, rng)
    
    # 3. Validate w/ Scipy
    quad_out, quad_err = integrate.quad(g, 0, np.inf)
    
    # 4. Compare
    print(f"\nMonte Carlo (Inversion): {mc_estimate:.8f}")
    print(f"SciPy quad (target):     {quad_out:.8f}")
    
    # The analytical solution is Gamma(2.5) + Gamma(1.5)
    # = (3/4)*sqrt(pi) + (1/2)*sqrt(pi) = (5/4)*sqrt(pi)
    analytical_sol = 1.25 * np.sqrt(np.pi)
    print(f"Analytical (target):     {analytical_sol:.8f}")

if __name__ == "__main__":
    main()
```

```python
# 7.252

import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# I_1 to I_3
def f1(x):
    return np.exp(np.sin(2*x))

def f2(x):
    return 1.0 / (2.0 + np.cos(x))

def f3(x):
    return np.exp(-x**2 / 2.0)

def plot_integrands():

    # 3 subplots for the 3 functions
    fig, (ax1, ax2, ax3) = plt.subplots(
        3, 1, 
        figsize=(7, 10), 
        tight_layout=True
    )
    fig.suptitle("Integrands for (7.252)", y=1.02)

    # Plot 1: I_1
    x1 = np.linspace(0, 2*np.pi, 400)
    y1 = f1(x1)
    ax1.plot(x1, y1, label=r'$f_1(x) = e^{\sin(2x)}$', color='blue')
    ax1.set_title(r'$I_1 = \int_0^{2\pi} e^{\sin(2x)} dx$')
    ax1.grid(True); ax1.legend()

    # Plot 2: I_2
    x2 = np.linspace(0, 2*np.pi, 400)
    y2 = f2(x2)
    ax2.plot(x2, y2, label=r'$f_2(x) = \frac{1}{2 + \cos x}$', color='green')
    ax2.set_title(r'$I_2 = \int_0^{2\pi} \frac{1}{2 + \cos x} dx$')
    ax2.grid(True); ax2.legend()

    # Plot 3: I_3
    x3 = np.linspace(-1, 1, 400)
    y3 = f3(x3)
    ax3.plot(x3, y3, label=r'$f_3(x) = e^{-x^2/2}$', color='red')
    ax3.set_title(r'$I_3 = \int_{-1}^1 e^{-x^2/2} dx$')
    ax3.grid(True); ax3.legend()
    
    plt.show()

def estimate_finite_integral(f, a, b, N_samples, rng):
    
    # 1. Draw N_samples from U(a, b)
    #    (This is the "inversion" x = a + (b-a)*u)
    x_samples = rng.uniform(a, b, size=N_samples)
    
    # 2. Calculate f(x) for all samples
    f_values = f(x_samples)
    
    # 3. The estimate is (width) * (average value)
    width = b - a
    estimate = width * np.mean(f_values)
    
    return estimate

def main():
    # 1. Plot
    plot_integrands()
    
    # 2. Integrate
    rng = np.random.default_rng()
    N_SAMPLES = 2_000_000
    
    print(f"\nRunning Monte Carlo with {N_SAMPLES:,} samples")

    # I_1
    a1, b1 = 0, 2*np.pi
    mc1 = estimate_finite_integral(f1, a1, b1, N_SAMPLES, rng)
    quad1, _ = integrate.quad(f1, a1, b1)
    print("\nIntegral I_1")
    print(f"Monte Carlo: {mc1:.8f}")
    print(f"SciPy quad (target):   {quad1:.8f}")

    # I_2
    a2, b2 = 0, 2*np.pi
    mc2 = estimate_finite_integral(f2, a2, b2, N_SAMPLES, rng)
    quad2, _ = integrate.quad(f2, a2, b2)
    print("\nIntegral I_2")
    print(f"Monte Carlo: {mc2:.8f}")
    print(f"SciPy quad (target):   {quad2:.8f}")
    
    # Analytical solution for I_2 is 2*pi/sqrt(3)
    analytical_sol_2 = (2 * np.pi) / np.sqrt(3)
    print(f"Analytical (target):   {analytical_sol_2:.8f}")

    # I_3
    a3, b3 = -1, 1
    mc3 = estimate_finite_integral(f3, a3, b3, N_SAMPLES, rng)
    quad3, _ = integrate.quad(f3, a3, b3)
    print("\nIntegral I_3")
    print(f"Monte Carlo: {mc3:.8f}")
    print(f"SciPy quad (target):   {quad3:.8f}")

if __name__ == "__main__":
    main()
```

```python
# 7.253

import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

def g(x):

    # np.maximum to avoid potential sqrt of negative
    x_safe = np.maximum(0, x)
    return np.sqrt(x_safe) * np.cos(x)

def plot_integrand():
    a, b = 0, 2*np.pi
    x_plot = np.linspace(a, b, 500)
    y_plot = g(x_plot)
    
    fig, ax = plt.subplots()
    ax.plot(x_plot, y_plot, label=r'$g(x) = \sqrt{x} \cos x$')
    ax.set_title("Integrand for (7.251)")
    ax.set_xlabel("x")
    ax.set_ylabel("g(x)")
    ax.axhline(0, color='black', linewidth=0.5)
    ax.grid(True)
    ax.legend()
    plt.show()
    
def run_monte_carlo(f, a, b, N_samples, rng):

    print(f"\nRunning Monte Carlo with {N_samples:,} samples")
    
    # 1. Draw N_samples from U(a, b)
    #    (This is the "inversion" x = a + (b-a)*u)
    x_samples = rng.uniform(a, b, size=N_samples)
    
    # 2. Calculate f(x) for all samples
    f_values = f(x_samples)
    
    # 3. The estimate is (width) * (average value)
    width = b - a
    estimate = width * np.mean(f_values)
    
    return estimate

def main():
    # 1. Plot
    plot_integrand()
    
    # 2. Integrate
    rng = np.random.default_rng()
    N_SAMPLES = 2_000_000
    a, b = 0, 2*np.pi
    
    mc_estimate = run_monte_carlo(g, a, b, N_SAMPLES, rng)
    
    # 3. Validate w/ Scipy
    quad_out, quad_err = integrate.quad(g, a, b)
    
    # 4. Compare
    print(f"\nMonte Carlo: {mc_estimate:.8f}")
    print(f"SciPy quad (target):   {quad_out:.8f}")

if __name__ == "__main__":
    main()
```
