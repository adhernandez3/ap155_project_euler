```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import time

en, f, f_sd = np.loadtxt("data_ch6_resonance.txt")

def breit_wigner(e, param_vec):
    e_center, gamma, strength = param_vec
    denom = ((e - e_center)**2) + ((gamma**2)/4) + 1e-15
    return strength/denom

def gauss_model(e, param_vec):
    e_center, gamma, strength = param_vec
    expon = ((e - e_center)**2)/(2*(gamma**2) + 1e-15)
    return strength*np.exp(-expon)

def chi_squared_bw(param_vec):
    if param_vec[1] <= 0 or param_vec[2] <= 0:
        return 1e100
    numer = f - breit_wigner(en, param_vec)
    denom = f_sd
    return np.sum((numer/denom)**2)

def chi_squared_gauss(param_vec):
    if param_vec[1] <= 0 or param_vec[2] <= 0:
        return 1e100
    numer = f - gauss_model(en, param_vec)
    denom = f_sd
    return np.sum((numer/denom)**2)

x0 = np.array([2.5, 0.1, 0.1])
print(f"Initial Guess (x0) = {x0}\n")
```
