```python
import matplotlib.pyplot as plt
import numpy as np

from decimal import Decimal
from fractions import Fraction

#float : imbes na gamitin lahat ng bit, gamitin yung float
# mantissa x exponent; mantissa yung sig fig
#PERO represented in binary yung mantissa and exp

# Classic floating point quirks example
print(0.1 + 0.2)
#pansinin, hindi siya tama dahil may 4 sa dulo

# How about increasing precision?
print(np.float128(0.1) + np.float128(0.2))
#pansinin, pwede na pero hindi pa rin

# How bad is decreasing precision?
print(np.float16(0.1) + np.float16(0.2))
#pansinin, MALI! really bad

# Closest we could get to fixed-point arithmetic: decimal
print(Decimal(0.1) + Decimal(0.2))
#pansinin, mas accurate

# How about using an explicit Fraction object type?
print(float(Fraction(1, 10) + Fraction(1, 5)))
#pansinin, pinaka tama PERO pinaka complex
```

```python
#Problem 2.6

a =1
b = 10**8
c =1

#a x_
#b x+ using bad formula
#c x+ using good formula

#a
x_neg = ((-b - np.sqrt(b**2 - 4*a*c))/(2*a))

#b
x_pos = ((-b + np.sqrt(b**2 - 4*a*c))/(2*a))

print(x_pos,x_neg)

#c
x_pos2 = c/(a *x_neg)
print(x_pos,x_pos2)
```
