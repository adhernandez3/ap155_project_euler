```python
import numpy as np
import matplotlib.pyplot as plt

def f_func(x):
    return((1-(np.cos(x)))/(x**2))

#b
def f_d2_central(x):
    h = np.logspace(-1, -6, 100)
    f_d2 = ( f_func(x+h) - 2*f_func(x) + f_func(x-h)) /(h**2)
    f_d2_anal = ((-4*(np.sin(x)))/(x**3))+((6*(1-np.cos(x)))/(x**4))+((np.cos(x))/(x**2))
    E = abs(f_d2 - f_d2_anal)
    return E

def f_d2_plot (x):
    h = np.logspace(-1, -6, 100)
    E = f_d2_central(x)
    fig, ax = plt.subplots()
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.plot(h,E)

f_d2_plot(0.004)

#c
def f_func_new(x):
    return ((2*((np.sin(x/2))**2)) / (x**2))

def central(x):
    h = np.logspace(-1, -6, 100)
    f_d2_c = ( f_func_new(x+h) - 2*f_func_new(x) + f_func_new(x-h)) /(h**2)
    f_d2_anal_c = ((-4*(np.sin(x)))/(x**3))+((6*(1-np.cos(x)))/(x**4))+((np.cos(x))/(x**2))
    O = abs(f_d2_c - f_d2_anal_c)
    return O

def f_d2_plot_c (x):
    h = np.logspace(-1, -6, 100)
    O = central(x)
    fig, ax = plt.subplots()
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.plot(h,O)

f_d2_plot_c(0.004)

#plt.savefig("pangalan") --> gamitin para masave yung plot to a file
```

```python
#3.25 a

import numpy as np
import matplotlib.pyplot as plt

def fnoise(x):
    return 2 + 5*np.sin(x) + 0.1*np.sin(30*x)

def gnoise(x):
    return 2 + 5*np.sin(x)

testx = np.linspace(0 , 2*np.pi, 128)

plt.plot(testx, fnoise(testx), label = r'f(x) =2 + 5sin(x) + 0.1sin(30x)')
plt.plot(testx, gnoise(testx), label = r'g(x) = 2 + 5sin(x)')
plt.legend()
plt.title("Plots of Noisy Functions")

plt.savefig("Plots of Noisy Functions")
```

```python
#3.25 b

def dfnoise(x):
    return 5*np.cos(x) + 0.1*30*np.cos(30*x)

def dgnoise(x):
    return 5*np.cos(x)

def num_dfnoise(x, h):
    return ((fnoise(x + h) - fnoise(x))/(h))

plt.plot(testx, dfnoise(testx), label = r"Analytical f'(x)")
plt.plot(testx, dgnoise(testx), label = r"Analytical g'(x)")
plt.plot(testx, num_dfnoise(testx,0.1), label = r"Numerical f'(X) for h = 0.1")
plt.plot(testx, num_dfnoise(testx,0.01), label = r"Numerical f'(X) for h = 0.01")
plt.legend()
plt.title('Analytical vs Numerical Differentiation (x = h)')

plt.savefig("Analytical vs Numerical Differentiation (x = h)")
```
