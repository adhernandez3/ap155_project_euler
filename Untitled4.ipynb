```python
import numpy as np
import matplotlib.pyplot as plt

#sinusoidal with varying phase shift

N = 5_000
x = np.linspace(0, 12*2*np.pi, N) 
deltaT = x[1] - x[0]

f = 1

#Original
y = np.cos(2*np.pi*f*x) #synthetic function: cos(2ft) where f = 1
Yk = np.fft.fft(y)
Yk_mag = np.abs(Yk)
Yk_phase = np.angle(Yk)
Yk_phase = np.unwrap(Yk_phase)

#Shift phase
y_shift = np.cos(2*np.pi*f*x + np.pi/2) 
Yk_shift = np.fft.fft(y_shift)
Yk_mag_shift = np.abs(Yk_shift)
Yk_phase_shift = np.angle(Yk_shift)
Yk_phase_shift = np.unwrap(Yk_phase_shift)

#Shift freq
y_shift_f = np.cos(2*np.pi*2*x) 
Yk_shift_f = np.fft.fft(y_shift_f)
Yk_mag_shift_f = np.abs(Yk_shift_f)
Yk_phase_shift_f = np.angle(Yk_shift_f)
Yk_phase_shift_f = np.unwrap(Yk_phase_shift_f)

freqs = np.fft.fftfreq(N, deltaT)

fig, ax = plt.subplots(1, 3, figsize=(12, 3))
plt.subplots_adjust(wspace=0.4)

#function plot
ax[0].plot(x, y, "k-", label="Cosine, f=1")
ax[0].plot(x, y_shift, "b-", label="Sine")
ax[0].plot(x, y_shift_f, "r-", label="Cosine, f=2")
ax[0].set_xlim(0, 4)
ax[0].set_title("Sinusoidal function with varying phase shift")
ax[0].set_xlabel("Time t")
ax[0].legend()

#transform plot
ax[1].plot(freqs, Yk_mag, "k.", label="Cosine, f=1")
ax[1].plot(freqs, Yk_mag_shift, "b.", label="Sine")
ax[1].plot(freqs, Yk_mag_shift_f, "r.", label="Cosine, f=2")
ax[1].set_xlim(0, 4)
ax[1].set_title("FT magnitude")
ax[1].set_xlabel("Frequency f")
ax[1].legend()

#phase plot
ax[2].plot(freqs, Yk_phase, "k.",label="Cosine, f=1")
ax[2].plot(freqs, Yk_phase_shift, "b.", label="Sine")
ax[2].plot(freqs, Yk_phase_shift_f, "r.", label="Cosine, f=2")
ax[2].set_title("FT phase")
ax[2].set_xlabel("Frequency f")
ax[2].legend()

for ax in ax:
    ax.grid(True)
fig.tight_layout() 
plt.savefig('sinusoid.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
N = 5000
x = np.linspace(0, 12 * 2 * np.pi, N)
deltaT = x[1] - x[0]

amplitude = 1.0
mean = 5
std_dev1 = 0.5
std_dev2 = 1.0
f_carrier = 2.0  # carrier frequency (wave packet center frequency)

# --- Gaussian wave packets ---
y1 = amplitude * np.exp(-((x - mean)**2) / (2 * std_dev1**2)) * np.cos(2 * np.pi * f_carrier * x)

# --- Fourier Transforms ---
Yk1 = np.fft.fftshift(np.fft.fft(y1))
freqs = np.fft.fftshift(np.fft.fftfreq(N, deltaT))

Yk_mag1 = np.abs(Yk1)

Yk_phase1 = np.unwrap(np.angle(Yk1))

# --- Plotting ---
fig, ax = plt.subplots(1, 3, figsize=(14, 4))
plt.subplots_adjust(wspace=0.4)

# Time-domain wave packets
ax[0].plot(x, y1, "k-", label=f"={std_dev1}")
ax[0].set_xlim(0, 10)
ax[0].set_title("Gaussian Wave Packets")
ax[0].set_xlabel("Time t")
ax[0].set_ylabel("Amplitude")
ax[0].legend()

# Fourier Transform Magnitude
ax[1].plot(freqs, Yk_mag1, "k.", label=f"={std_dev1}")
ax[1].set_xlim(-5, 5)
ax[1].set_title("FT Magnitude")
ax[1].set_xlabel("Frequency f")
ax[1].legend()

# Fourier Transform Phase
ax[2].plot(freqs, Yk_phase1, "k-", label=f"={std_dev1}")
ax[2].set_xlim(-5, 5)
ax[2].set_title("FT Phase")
ax[2].set_xlabel("Frequency f")
ax[2].legend()

for ax in ax:
    ax.grid(True)
fig.tight_layout() 
plt.savefig('gaussian wave packet.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
N = 5000
x = np.linspace(0, 12 * 2 * np.pi, N)
deltaT = x[1] - x[0]

amplitude = 1.0
mean = 5
std_dev1 = 0.5
std_dev2 = 1.0
f_carrier = 2.0  # carrier frequency (wave packet center frequency)

# --- Gaussian wave packets ---
y1 = amplitude * np.exp(-((x - mean)**2) / (2 * std_dev1**2)) * np.cos(2 * np.pi * f_carrier * x)
y2 = amplitude * np.exp(-((x - mean)**2) / (2 * std_dev2**2)) * np.cos(2 * np.pi * f_carrier * x)

# --- Fourier Transforms ---
Yk1 = np.fft.fftshift(np.fft.fft(y1))
Yk2 = np.fft.fftshift(np.fft.fft(y2))
freqs = np.fft.fftshift(np.fft.fftfreq(N, deltaT))

Yk_mag1 = np.abs(Yk1)
Yk_mag2 = np.abs(Yk2)

Yk_phase1 = np.unwrap(np.angle(Yk1))
Yk_phase2 = np.unwrap(np.angle(Yk2))

# --- Plotting ---
fig, ax = plt.subplots(1, 3, figsize=(14, 4))
plt.subplots_adjust(wspace=0.4)

# Time-domain wave packets
ax[0].plot(x, y1, "b-", label=f"={std_dev1}")
ax[0].plot(x, y2, "k-", label=f"={std_dev2}")
ax[0].set_xlim(0, 10)
ax[0].set_title("Gaussian Wave Packets")
ax[0].set_xlabel("Time t")
ax[0].set_ylabel("Amplitude")
ax[0].legend()

# Fourier Transform Magnitude
ax[1].plot(freqs, Yk_mag1, "b.", label=f"={std_dev1}")
ax[1].plot(freqs, Yk_mag2, "k.", label=f"={std_dev2}")
ax[1].set_xlim(-5, 5)
ax[1].set_title("FT Magnitude")
ax[1].set_xlabel("Frequency f")
ax[1].legend()

# Fourier Transform Phase
ax[2].plot(freqs, Yk_phase1, "b.", label=f"={std_dev1}")
ax[2].plot(freqs, Yk_phase2, "k.", label=f"={std_dev2}")
ax[2].set_xlim(-5, 5)
ax[2].set_title("FT Phase")
ax[2].set_xlabel("Frequency f")
ax[2].legend()

fig.tight_layout()
plt.show()
```
