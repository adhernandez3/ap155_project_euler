```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

# ============================================================
# 1. LOAD & PARSE YOUR DATA FILE (MM:SS.s timestamp format)
# ============================================================

times = []
intensities = []

with open("data.txt", "r") as f:
    for line in f:
        line = line.strip()
        if line == "":
            continue

        parts = line.split()
        if len(parts) < 2:
            continue

        t_str = parts[0]
        i_str = parts[-1]

        # Parse intensity
        try:
            intensity = float(i_str)
        except:
            continue

        # Parse timestamp in MM:SS.s format
        try:
            t_obj = datetime.strptime(t_str, "%M:%S.%f")
        except:
            continue

        times.append(t_obj)
        intensities.append(intensity)

intensities = np.array(intensities)

# ============================================================
# 2. CONVERT TIMESTAMPS TO SECONDS (RELATIVE TIME AXIS)
# ============================================================

reference = times[0]
time_sec = np.array([(t - reference).total_seconds() for t in times])

# Sort (just in case)
order = np.argsort(time_sec)
time_sec = time_sec[order]
intensities = intensities[order]

# Estimate sampling rate
dt = np.mean(np.diff(time_sec))
Fs = 1 / dt

print(f"\nEstimated sampling interval dt = {dt:.6f} seconds")
print(f"Estimated sampling rate Fs â‰ˆ {Fs:.2f} Hz")


# ============================================================
# 3. FFT COMPUTATION
# ============================================================

N = len(intensities)
Yk = np.fft.fft(intensities)
freqs = np.fft.fftfreq(N, dt)

Yk_mag = np.abs(Yk) / N
Yk_phase = np.unwrap(np.angle(Yk))

# Positive frequencies only
pos = freqs >= 0
freqs_pos = freqs[pos]
Yk_mag_pos = Yk_mag[pos]


# ============================================================
# 4. AUTOMATIC PEAK FREQUENCY DETECTION
# ============================================================

# Remove DC component
mask_nonzero = freqs_pos > 0
freqs_nz = freqs_pos[mask_nonzero]
mag_nz = Yk_mag_pos[mask_nonzero]

# Find strongest peak
peak_idx = np.argmax(mag_nz)
peak_freq = freqs_nz[peak_idx]
peak_amp = mag_nz[peak_idx]

print(f"\nDetected LED modulation frequency: {peak_freq:.3f} Hz")
print(f"Peak amplitude: {peak_amp:.3f}")


# ============================================================
# 5. ZOOM FFT AROUND PEAK
# ============================================================

zoom = 10.0  # Â±10 Hz window
zoom_mask = (freqs_pos > (peak_freq - zoom)) & (freqs_pos < (peak_freq + zoom))


plt.figure(figsize=(8,4))
plt.plot(freqs_pos[zoom_mask], Yk_mag_pos[zoom_mask], 'm-')
plt.title(f"Zoomed FFT around {peak_freq:.2f} Hz")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()


# ============================================================
# 6. BAND-PASS FILTERING AROUND PEAK FREQUENCY
# ============================================================

band = 5.0  # keep Â±5 Hz around the detected frequency

Y_filtered = Yk.copy()
remove_mask = np.abs(freqs - peak_freq) > band
Y_filtered[remove_mask] = 0

# Inverse FFT to recover band-passed time signal
filtered_signal = np.fft.ifft(Y_filtered).real


plt.figure(figsize=(10,3))
plt.plot(time_sec, filtered_signal, 'g-')
plt.title("Band-Passed Signal (Â±5 Hz around LED frequency)")
plt.xlabel("Time (s)")
plt.grid(True)
plt.show()


# ============================================================
# 7. PYTHON LOCK-IN AMPLIFIER (SYNCHRONOUS DEMODULATION)
# ============================================================

# Reference oscillator
reference_signal = np.sin(2 * np.pi * peak_freq * time_sec)

# Mix
mixed_signal = intensities * reference_signal

# Moving average (LPF)
window = int(0.5 * Fs)     # 0.5 second smoothing
kernel = np.ones(window) / window
lockin_output = np.convolve(mixed_signal, kernel, mode='same')

plt.figure(figsize=(10,3))
plt.plot(time_sec, lockin_output, 'r-')
plt.title("Lock-in Demodulated Amplitude (Python)")
plt.xlabel("Time (s)")
plt.grid(True)
plt.show()


# ============================================================
# 8. SUMMARY PLOTS (OPTIONAL)
# ============================================================

fig, ax = plt.subplots(1, 3, figsize=(15,4))
plt.subplots_adjust(wspace=0.3)

# Raw signal
ax[0].plot(time_sec, intensities, 'k-')
ax[0].set_title("Raw Intensity")
ax[0].set_xlabel("Time (s)")
ax[0].grid(True)

# Full FFT magnitude
ax[1].plot(freqs_pos, Yk_mag_pos, 'b-')
ax[1].set_title("FFT Magnitude")
ax[1].set_xlabel("Frequency (Hz)")
ax[1].grid(True)

# Lock-in output
ax[2].plot(time_sec, lockin_output, 'r-')
ax[2].set_title("Lock-in Output")
ax[2].set_xlabel("Time (s)")
ax[2].grid(True)

plt.show()
```
