```python
t = np.arange(0, 10, 0.05)
freq = 1.0

y=np.sin(2*np.pi*freq*t)

fig, ax = plt.subplots()

shifted_y = np.sin(2*np.pi*freq*t - 2.0)

ax.plot(t, y, label = "not shifted")
ax.plot(t, shifted_y, label="is shifted")
ax.legend()
```

```python
#Exercise 1

t = np.arange(0, 5, 0.01)
freq_1 = 1.0
freq_2 = 10.0

y_1 = np.sin(2*np.pi*freq_1*t)
y_2 = np.sin(2*np.pi*freq_2*t)

fig, ax = plt.subplots()

ax.plot(t, y_1*y_2, label = "product")
ax.plot(t, y_1)
ax.legend()
```

```python
#Exercise 2

t=np.arange(0,1,0.00005)

y = np.sqrt(1 - t**2)

fig, ax = plt.subplots()

counter_in=0
counter_out=0

for i in range(1000):
    x_i = np.random.random(10000)
    y_i = np.random.random(10000)
    if (x_i**2 + y_i**2 <= 1):
        ax.scatter(x_i, y_i, color="blue")
        counter_in +=1
        
    else:
        ax.scatter(x_i, y_i, color="red")
        counter_out +=1

plt.plot(t,y)
print (counter_in)
print (counter_out)
print (4*(counter_in/(counter_in+counter_out)))
```

```python
N=1000000
x_i = np.random.random(size = N)
y_i = np.random.random(size = N)

is_inside = x_i**2 + y_i**2 < 1

print(is_inside)

fig, ax = plt.subplots()

ax.scatter(x_i[is_inside], y_i[is_inside])

4*np.sum((is_inside)/N)
```
