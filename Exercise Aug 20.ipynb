```python
#A.1. Create a function random_integer_matrix(nrows, ncols) that generates a matrix in the 
# form of "list-of-lists" whose elements are random integeres from 1 to 9. 
# Each "sub-list" inside the main list is a row in the matrix.

import random
# Docs: https://docs.python.org/3/library/random.html

def random_integer_matrix(nrows, ncols):
    output_matrix = []

    for col in range(ncols):
        new_col = []
        for row in range(nrows):
            pick_number = random.randint(1, 9)
            new_col.append(pick_number)
        output_matrix.append(new_col)
    return output_matrix


random_integer_matrix(5,4)
```

```python
# A.2. Create a function scalar_product(list1, list2) where you multiply each element 
# at the same index in the two lists and sum/accumulate the results.

def scalar_product(list1, list2):
    total_product=0
    for i in range(len(list1)):
        total_product+=list1[i] * list2[i]
    return total_product


scalar_product([1,2,3], [4,5,6])
```
